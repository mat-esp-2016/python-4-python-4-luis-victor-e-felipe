```python
def nome_da_cidade (nome):
    data = open(nome, encoding = "UTF-8")
    x = data.readlines()
    coord = x [4] [1:]
    nomes = x [35].split(sep = ',')[0][18:] 

    
    return (coord, nomes)
# Definimos uma função para ler as linhas na função x
#coord entramos com os valores das coordenadas com as strings 
#nomes separamos as strings com.split e sep =','para separar as cidades como string e usamos só os nomes das cidades
```

### Tarefa 4 

```python
def temperaturas_estimada (estacao):
    data = open(estacao) 
    coluna = np.loadtxt(estacao, comments = '%')
    coluna_1 = np.nan_to_num (coluna[0:,4])
    x = data.readlines()
    y = x[47].split(sep = ' ')
    z = float (y[-3])
    a = np.ones(len(coluna_1))
    h = z*a 
    final = h + coluna_1
    
    
    return final

# primeiro definimos a função temperaturas para fazer as contas esperadas e colocamos a variável estação
# data foi usada para abrir o arquivo da variável estação
# coluna importamos somente os dados de número e coluna_1 convertemos nan para 0, e pegamos a coluna referente a anomalia anual
# x lê as linhas de data que foi o arquivo todo aberto
# y foi para pegar a linha 47 e separamos as strings por espaço
# z transformamos o numero da temperatura escolhida de uma string para um float
# a é uma colina criada com vários numero 1 do tamanho da coluna de anomalia anual
# h pegamos o float criado na variável z e multiplicamos por toda a coluna de números 1
# final é o resultado final entre a soma das colunas de temperatura com as anomalias anuais
```

### Tarefa 5

```python
def meses_e_anos_para_anos_decimais (local):
    coluna = np.loadtxt(local, comments = '%')
    anos = coluna[:,0]
    meses = coluna[:,1]
    anos_decimal = anos + 1/12 * (meses-1)
    return anos_decimal

# Definimos a função para o cálculo dos meses e anos para anos decimais
```

```python
def pratica_2_graficos_plotados (imagem):
    ano = meses_e_anos_para_anos_decimais(imagem)
    temp = temperaturas_estimada(imagem)
    
    plt.figure()
    plt.plot (ano,temp,'r-')
    plt.xlabel ('Anos')
    plt.ylabel ('Temperatura anual estimada(°C) ')
    plt.title (nome_da_cidade(imagem)[0])
    plt.savefig(nome_da_cidade(imagem)[1] +'.png')
    return 

# para plotar e salvar o gráfico em png foi criada uma função onde a variável ano veio da função já criada que nos dá o ano decimal
# e a variável temp nos dá a temperatura calculada da função temperatura_estimada
# logo após plotamos os gráficos 
# para o título ser igual ao nome da cidade, foi usada a função nome_da_cidade com a segunda string, que é referente ao nome da cidade
# como precisamos salvar a figura, usamos o plt.savefig com o nome da cidade que foi dado no título adicionando o formato .png
```
